```python
import numpy as np

class Neuron:
    def __init__(self, input_vector, weights):
        self.input_vector = input_vector
        self.weights = weights


    def add_function(self):
      #Compute the weighted sum of the input vector elements multiplied by the corresponding weights.

        total = 0
        for i in range(len(self.input_vector)):
            total += self.input_vector[i] * self.weights[i]
        return total

    def ReLU(self, input_value):
        if input_value > 0:
            return input_value
        else:
            return 0

input_vector = [1, 3, 5, 6, 9]
weights = [2, 4, 6, 7, 8]

# Create three neurons with the same input vector and weights
neuron_1 = Neuron(input_vector, weights)
neuron_2 = Neuron(input_vector, weights)
neuron_3 = Neuron(input_vector, weights)

# Compute output for each neuron
output_1 = neuron_1.add_function()
output_2 = neuron_2.add_function()
output_3 = neuron_3.add_function()

# Print neuron outputs
print("Neuron 1 output:", output_1)
print("Neuron 2 output:", output_2)
print("Neuron 3 output:", output_3)
```

```python
# Create a new neuron with the outputs of the first three neurons as input

input_vector_output = [output_1, output_2, output_3]
random_weights_output = np.random.uniform(0, 3, size=3)
print("Random weights:", random_weights_output)
neuron_output = Neuron(input_vector_output, random_weights_output)

# Compute output for the neurons
output = neuron_output.add_function()

print("Output neuron output:", output)
```
